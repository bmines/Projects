```python
import pandas as pd
data = pd.read_csv("/content/CC GENERAL.csv")
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data = data.drop("CUST_ID", axis = 1)
data
```

```python
data.fillna(data["CREDIT_LIMIT"].mean(), inplace = True)
data.fillna(data["MINIMUM_PAYMENTS"].mean(), inplace = True)
data
```

```python
data.isnull().sum()
```

```python
data.isna().sum()
```

```python
data = data + 0.0
```

```python
data.dtypes
```

```python
x = data[["BALANCE", "PURCHASES"]]
x
```

```python
#standardized x
from sklearn import preprocessing
standard = preprocessing.scale(x)
print(standard)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 4, affinity = "euclidean", linkage = "complete")
clust_labels = model.fit_predict(standard)
```

```python
agglomerative = pd.DataFrame(clust_labels)
agglomerative[0].value_counts()
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(standard[:,0], standard[:,1], c = agglomerative[0], s = 50)
plt.colorbar(scatter)
```

```python
from sklearn.cluster import KMeans 
kmeans=KMeans(n_clusters=5, random_state=0) 
kmeans.fit(standard)
```

```python
labels = pd.DataFrame(kmeans.labels_)
df=pd.DataFrame(standard)
df2 =  pd.concat([labels, df], axis=1)
df2.columns = ["labels", "BALANCE", "PURCHASES"]
df2
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
dend = shc.dendrogram(shc.linkage(data, method = "complete"))
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters =i, init="k-means++", max_iter=300, n_init=10)
    kmeans.fit(df2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss);
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans.predict(standard)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(df2["BALANCE"][df2.labels == 0],
            df2["PURCHASES"][df2.labels == 0],s=80,c='magenta',label='Careful')
plt.scatter(df2["BALANCE"][df2.labels == 1],
           df2["PURCHASES"][df2.labels == 1],s=80,c='yellow',label='Standard')
plt.scatter(df2["BALANCE"][df2.labels == 2],
           df2["PURCHASES"][df2.labels == 2],s=80,c='green',label='Target')
plt.scatter(df2["BALANCE"][df2.labels == 3], 
           df2["PURCHASES"][df2.labels == 3],s=80,c='cyan',label='Careless')
plt.scatter(df2["BALANCE"][df2.labels == 4], 
           df2["PURCHASES"][df2.labels == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.legend()
plt.show()
```
