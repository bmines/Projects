```python
import pandas as pd
data = pd.read_csv("/content/CC GENERAL.csv")
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data = data.drop("CUST_ID", axis = 1)
data
```

```python
data.fillna(data["CREDIT_LIMIT"].mean(), inplace = True)
data.fillna(data["MINIMUM_PAYMENTS"].mean(), inplace = True)
data
```

```python
data.isnull().sum()
```

```python
data.isna().sum()
```

```python
data = data + 0.0
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import StandardScaler, normalize
scaler = StandardScaler()
x_scaled = scaler.fit_transform(data)
standard = normalize(x_scaled)
print(standard)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_principal = pca.fit_transform(standard)
x_principal = pd.DataFrame(x_principal)
x_principal.columns = ["P1", "P2"]
x_principal
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 4, affinity = "euclidean", linkage = "complete")
clust_labels = model.fit_predict(x_principal)
```

```python
agglomerative = pd.DataFrame(clust_labels)
agglomerative
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
ac5 = AgglomerativeClustering(n_clusters = 5)
plt.figure(figsize =(6, 6))
plt.scatter(x_principal['P1'], x_principal['P2'], 
           c = ac5.fit_predict(x_principal), cmap ='rainbow')
plt.show()
```

```python
from sklearn.cluster import KMeans 
kmeans=KMeans(n_clusters=5, random_state=0) 
kmeans.fit(x_principal)
```

```python
labels = pd.DataFrame(kmeans.labels_)
df=pd.DataFrame(x_principal)
df2 =  pd.concat([labels, x_principal], axis=1)
df2.columns = ["labels", "BALANCE", "PURCHASES"]
df2
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(8, 8))
dend = shc.dendrogram(shc.linkage(x_principal, method = "ward"))
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters =i, init="k-means++", max_iter=300, n_init=10)
    kmeans.fit(df2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss);
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```
